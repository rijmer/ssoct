```python
RAW_URL = "https://raw.githubusercontent.com/kirvis13/weheat-log-fetcher/main/log-fetcher.py"
import requests, os

print("Downloading script from:", RAW_URL)
r = requests.get(RAW_URL, timeout=30)
r.raise_for_status()
with open("log-fetcher.py", "w", encoding="utf-8") as f:
    f.write(r.text)
print("âœ… Script downloaded. Size:", len(r.text), "bytes")


```
